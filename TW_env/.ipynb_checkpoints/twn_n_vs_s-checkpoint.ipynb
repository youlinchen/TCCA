```python
def split_ws(X):
    (T, P, S)= X.shape
    W = np.zeros((int(T/2), P, S))
    S = np.zeros((int(T/2), P, S))
    w = 0
    s = 0
    for t in range(T):
        if (t%12<3)|(t%12>8):
            W[w,:,:] = X[t, :,:]
            w += 1
        else:
            S[s,:,:] = X[t, :,:]
            s += 1
    return W, S
```

```python
X_W, X_S = split_ws(TW_n)
Y_W, Y_S = split_ws(TW_s)
```

```python
inputX, inputY = X_W, Y_W

(loading, corr, error) = twoDcca(inputX, inputY, x_regular=0.01, y_regular=0.01, iter_max=30)
for i in range(20):
    (loading_tmp, corr_tmp, error_tmp) = twoDcca(inputX, inputY, x_regular=0.01, y_regular=0.01, iter_max=30)
    if corr[-1]<corr_tmp[-1]:
        corr = corr_tmp
        loading = loading_tmp
print(corr[-1])
```
